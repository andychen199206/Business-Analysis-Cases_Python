```python
import numpy as np
#print(df130202rs[((df130202rs['ENTRIES'] <10000) | (df130202rs['EXITS']<10000)) & (df130202rs['DATE']=='02-01-13') & (df130202rs['DESC']=='REGULAR')])

df130201=df130202rs[(df130202rs['DATE']=='02-01-13')]
#print(df130201.count() )  #30591 rows no NaN
#print(df130201[df130201['DESC']=='REGULAR'].count() ) #27093 rows no NaN


df130131=df130202rs[df130202rs['DATE']=='01-31-13']
#print(df130131.count())#33873 rows no NaN
#print(df130131[df130131['DESC']=='REGULAR'].count() ) #27353 rows no NaN

#print(df130201.groupby(['C/A', 'UNIT', 'SCP']).agg({'TIME': ['min','max']}))

dfcb=pd.merge(df130201, df130131,how='inner', on=['C/A','UNIT','SCP','TIME'])

#print(dfcb.count()) #27608 rows no NaN

dfcbMaxTime=dfcb.groupby(['C/A', 'UNIT', 'SCP']).agg({'TIME': 'max'}).reset_index()

#print( dfcbMaxTime['TIME'].nunique()) #15 unique values

EE=pd.merge(dfcb, dfcbMaxTime, how='inner', on=['C/A','UNIT','SCP','TIME'])

EE['ENTRIESDelta']=EE['ENTRIES_x']-EE['ENTRIES_y']
EE['EXITSDelta']=EE['EXITS_x']-EE['EXITS_y']

print(EE['ENTRIESDelta'].sum())

print(EE['EXITSDelta'].sum())

print(EE[(EE['ENTRIESDelta']<0) | (EE['EXITSDelta']<0)])
```

 Let’s define the busy-ness as sum of entry & exit count. What station was the busiest on February 1, 2013? What turnstile was the busiest on that date?

```python
EE['E+E']=EE['ENTRIESDelta']+EE['EXITSDelta']

print(EE['E+E'].groupby(EE['UNIT']).sum())
```

What stations have seen the most usage growth/decline in 2013?

```python
from datetime import datetime, timedelta
#import pandas as pd
date=datetime(2013,1,5)
df2013=pd.DataFrame([])
    

while date<=datetime(2014,1,4):
    datestr=date.strftime("%y%m%d")
    mtaurl='http://web.mta.info/developers/data/nyct/turnstile/turnstile_' + datestr + '.txt'
    #
    date=date+timedelta(weeks=1)
    try:
        dfnew= pd.read_csv(mtaurl, header=None)
        df2013=df2013.append(dfnew)
    except Exception:
        print(mtaurl) #print the url if it wasn't load correctly
        pass
df2013.columns=['C/A','UNIT','SCP','DATE1','TIME1','DESC1','ENTRIES1','EXITS1','DATE2','TIME2','DESC2','ENTRIES2','EXITS2',
                  'DATE3','TIME3','DESC3','ENTRIES3','EXITS3','DATE4','TIME4','DESC4','ENTRIES4','EXITS4',
                  'DATE5','TIME5','DESC5','ENTRIES5','EXITS5','DATE6','TIME6','DESC6','ENTRIES6','EXITS6',
                  'DATE7','TIME7','DESC7','ENTRIES7','EXITS7','DATE8','TIME8','DESC8','ENTRIES8','EXITS8']
print(df2013.head())
```

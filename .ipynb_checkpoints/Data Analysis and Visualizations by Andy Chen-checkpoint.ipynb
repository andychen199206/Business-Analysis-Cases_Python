#  Data Analysis and Visualizations by Andy Chen

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
df = pd.read_excel('E:\Data Test\Campfire Analytics\CampfireAnalytics_Candidate_DataSet.xlsx', sheet_name='DataSet', skiprows=2)
```

# 1. Which traffic sources perform best, and why?
Answer: Affiliate has the best performance because Affiliate has the highest Conversion Rate(Purchases/Total Visits), even though Direct Traffic has the highest number of Total Visits and the highest number of Purchases. We need to focus on Conversion Rate which means efficiency, so we can invest more in Affiliate Marketing.

```python
df1=df.groupby(['Referral Source'])['Visits','Bounced Visits','Total Visits','Checkouts','Purchases'].sum()
print(df1)
df1['Total Visits'].sort_values(ascending = True).plot.barh(color='Blue')
plt.title('Traffic Performance based on Number of Total Visits')
plt.xlabel('Number of Total Visits')
plt.xticks(rotation=30)
```

```python
df1['Purchases'].sort_values(ascending = True).plot.barh(color='Red')
plt.title('Traffic Performance based on Number of Purchases')
plt.xlabel('Number of Purchases')
```

```python
df1['Purchases/Total Visits']=df1['Purchases']/df1['Total Visits']
df1['Purchases/Total Visits'].sort_values(ascending=True).plot.barh(color='Orange')
plt.title('Traffic Performance based on Conversion Rate(Purchases/Total Visits)')
plt.xlabel('Conversion Rate(Purchases/Total Visits)')
```

# 2. Do there appear to be checkout funnel issues for any particular countries or traffic sources? Please explain.
Answer: Paid Social, China. Paid Social has checkout funnel issues because its rate of 'Checkouts/Total Visits' and 'Purchases/Checkouts' are all lowest, which means very few percent of visitors from paid social go to checkouts and very few percent of visitors in checkouts will complete a purchase. The same reason for China.

```python
df_source=df.groupby(['Referral Source'])['Visits','Bounced Visits','Total Visits','Checkouts','Purchases'].sum()
print(df_source)
df_country=df.groupby(['Country'])['Visits','Bounced Visits','Total Visits','Checkouts','Purchases'].sum()
print(df_country)
```

```python
df_source['Bounced Visits/Total Visits']=df_source['Bounced Visits']/df_source['Total Visits']
df_source['Checkouts/Total Visits']=df_source['Checkouts']/df_source['Total Visits']
df_source['Purchases/Total Visits']=df_source['Purchases']/df_source['Total Visits']
df_source['Purchases/Checkouts']=df_source['Purchases']/df_source['Checkouts']
df_source[['Checkouts/Total Visits','Purchases/Checkouts']].sort_values('Checkouts/Total Visits').plot.bar(alpha=0.8)
plt.xticks(rotation=30)
plt.ylabel('Checkouts/Total Visits and Purchases/Checkouts rate')
plt.title('Performance of Checkout Funnel group by Traffic Sources')
```

```python
df_country['Bounced Visits/Total Visits']=df_country['Bounced Visits']/df_country['Total Visits']
df_country['Checkouts/Total Visits']=df_country['Checkouts']/df_country['Total Visits']
df_country['Purchases/Total Visits']=df_country['Purchases']/df_country['Total Visits']
df_country['Purchases/Checkouts']=df_country['Purchases']/df_country['Checkouts']
df_country[['Checkouts/Total Visits','Purchases/Checkouts']].sort_values('Checkouts/Total Visits').plot.bar(alpha=0.8)
plt.xticks(rotation=30)
plt.ylabel('Checkouts/Total Visits and Purchases/Checkouts rate')
plt.title('Performance of Checkout Funnel group by Countries')
```

# 3. Do there appear to be any seasonal trends that we should be aware of?
Answer: Yes. The number of total visits has a decrease in Nov and Feb,but increase in Dec and Jan.

```python
df['Date']=pd.to_datetime(df['Year'].astype(str)+df['Month'], format='%Y%b')
df_seasonal=df.groupby(['Date'])['Visits','Bounced Visits','Total Visits','Checkouts','Purchases'].sum()
print(df_seasonal)
```

```python
df_seasonal.plot()
plt.legend(bbox_to_anchor=(1.0, 1.02))
plt.title('Seasonal Trends')
plt.ylabel('Numbers')
plt.xlabel('Dates')
```

# 4. Which pages are most effective, and why?
Answer: Product detail pages,Homepage and Product List A. Because their conversion rate are the highest.

```python
df_pages=df.groupby(['Page'])['Visits','Bounced Visits','Total Visits','Checkouts','Purchases'].sum()
df_pages['Purchases/Total Visits']=df_pages['Purchases']/df_pages['Total Visits']
df_pages['Purchases/Total Visits'].sort_values(ascending = True).plot.barh(alpha=0.8,color='Blue')
plt.xlabel('Conversation Rate(Purchases/Total Visits)')
plt.title('Conversion Rate of each page')
```

# 5. What else can you tell us from this data that may help inform future business decisions or optimizations? 
Answer: Promotional Page has the highest Bounce Rate. Paid Display has the highest Bounce Rate. And the Bounce Rate is highest in Germany. Conversion Rate in highest in United States.

```python
df_pages['Bounced Visits/Total Visits']=df_pages['Bounced Visits']/df_pages['Total Visits']
df_pages['Bounced Visits/Total Visits'].sort_values(ascending = True).plot.barh(alpha=0.8,color='Blue')
plt.xlabel('Bounce Rate(Bounced Visits/Total Visits)')
plt.title('Bounce Rate of each page')
```

```python
df_source['Bounced Visits/Total Visits'].sort_values(ascending = True).plot.barh(alpha=0.8,color='Red')
plt.xlabel('Bounce Rate(Bounced Visits/Total Visits)')
plt.title('Bounce Rate of each Traffic Source')
```

```python
df_country['Bounced Visits/Total Visits'].sort_values(ascending = True).plot.barh(alpha=0.8,color='Orange')
plt.xlabel('Bounce Rate(Bounced Visits/Total Visits)')
plt.title('Bounce Rate of each Country')
```

```python
df_country[['Purchases/Total Visits']].sort_values('Purchases/Total Visits').plot.bar(alpha=0.8)
plt.xticks(rotation=30)
plt.ylabel('Conversion Rate(Purchases/Total Visits)')
plt.title('Conversion Rate of each Country')
```

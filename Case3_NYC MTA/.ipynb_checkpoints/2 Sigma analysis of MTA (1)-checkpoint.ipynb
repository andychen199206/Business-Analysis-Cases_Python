NYC MTA Turnstile Usage Dataset

Have a look at / download the following dataset :

- Go to http://web.mta.info/developers/turnstile.html

- This dataset shows entry & exit counter values for each turnstile-device in each station in the NYC Subway System.

- Note these aren’t counts per interval, but equivalent to an “odometer” reading for each device.

Which station has the most number of units?

Station: 23 ST

```python
df180505=pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180505.txt')
print(df180505.groupby('STATION').UNIT.nunique().reset_index(name='count').sort_values(['count'],ascending=False).head(1))
df180428=pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180428.txt')
print(df180428.groupby('STATION').UNIT.nunique().reset_index(name='count').sort_values(['count'],ascending=False).head(1))
```

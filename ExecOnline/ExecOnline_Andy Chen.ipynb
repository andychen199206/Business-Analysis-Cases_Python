```python
import pandas as pd
import matplotlib.pyplot as pt
import matplotlib.patches as mp
df = pd.read_excel('ExecOnline_dataset.xlsx')

#1.Reflecting on your participation in the program, how would you rate the value of your experience?
df['Reflecting on your participation in the program, how would you rate the value of your experience?'].value_counts().plot(kind = 'barh')
pt.title('Rate on program')
pt.xlabel('Frequency')
```
